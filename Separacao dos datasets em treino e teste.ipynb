```python
import os
import cv2
import numpy as np
import natsort
import os
from pathlib import Path
import shutil
import pandas as pd
```

```python
datasets=['circle_hilbert/circle_hilbert.json','trapezoid_hilbert/trapezoid_hilbert.json',
          'c_sector_hilbert/c_sector_hilbert.json','half_circle_hilbert/half_circle_hilbert.json']

for dataset in datasets[:]:
    df = pd.read_json(dataset)
    trainvalboxes=75
    trainvalpieces=75*51
    dftrainval=df[:trainvalpieces]
    dftest=df[trainvalpieces:]

    trainboxes=50
    trainpieces=50*51
    dftrain=dftrainval[:trainpieces]
    dfval=dftrainval[trainpieces:]
    
    trainlist=[]
    for index, row in dftrain.iterrows():
        references={}
        path,file=os.path.split(row['file'])
        path=path.replace('_hilbert','NN/train')
        Path(path).mkdir(parents=True, exist_ok=True)
        savename=os.path.join(path,file)
        references['file']=savename
        references['Y']=row['Y']
        trainlist.append(references)
        shutil.copyfile(row['file'],savename)
        #print(savename)
    nome=os.path.split(os.path.split(path)[0])[0]   
    newdftrain = pd.DataFrame.from_dict(trainlist)
    newdftrain.to_json(os.path.join(nome,'train'+'.json'))
    
    vallist=[]
    for index, row in dfval.iterrows():
        references={}
        path,file=os.path.split(row['file'])
        path=path.replace('_hilbert','NN/val')
        Path(path).mkdir(parents=True, exist_ok=True)
        savename=os.path.join(path,file)
        references['file']=savename
        references['Y']=row['Y']
        vallist.append(references)
        shutil.copyfile(row['file'],savename)
        #print(savename)
    nome=os.path.split(os.path.split(path)[0])[0]   
    newdfval = pd.DataFrame.from_dict(vallist)
    newdfval.to_json(os.path.join(nome,'val'+'.json'))
    
    trainvallist=[]
    for index, row in dftrainval.iterrows():
        references={}
        path,file=os.path.split(row['file'])
        path=path.replace('_hilbert','NN/trainval')
        Path(path).mkdir(parents=True, exist_ok=True)
        savename=os.path.join(path,file)
        references['file']=savename
        references['Y']=row['Y']
        trainvallist.append(references)
        shutil.copyfile(row['file'],savename)
        #print(savename)
    nome=os.path.split(os.path.split(path)[0])[0]   
    newdftrainval = pd.DataFrame.from_dict(trainvallist)
    newdftrainval.to_json(os.path.join(nome,'trainval'+'.json'))
    
    testlist=[]
    for index, row in dftest.iterrows():
        references={}
        path,file=os.path.split(row['file'])
        path=path.replace('_hilbert','NN/test')
        Path(path).mkdir(parents=True, exist_ok=True)
        savename=os.path.join(path,file)
        references['file']=savename
        references['Y']=row['Y']
        testlist.append(references)
        shutil.copyfile(row['file'],savename)
        #print(savename)
    nome=os.path.split(os.path.split(path)[0])[0]   
    newdftest = pd.DataFrame.from_dict(testlist)
    newdftest.to_json(os.path.join(nome,'test'+'.json'))
    



    
    
    
    
    


```
